```python
import pandas as pd
from functools import reduce
```

```python
table01 = "tables/table01.csv"
table02 = "tables/table02a.csv"
table03 = "tables/table02b.csv"
table04 = "tables/table02c.csv"
table05 = "tables/table02d.csv"
```

```python
table01_df = pd.read_csv(table01)
table02_df = pd.read_csv(table02)
table03_df = pd.read_csv(table03)
table04_df = pd.read_csv(table04)
table05_df = pd.read_csv(table05)
```
