```python
#Dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import time
```

```python
order = ['1st','2nd','3rd']
for i in range(4,17):
    item = str(i)+"th"
    order.append(item)
print(order)
```

```python
data = pd.DataFrame(columns={"Name":"","Party":"","Receipts*":"","Disbursements**":"","Cash on hand":"","Date":"","District":""},index=[0])


for i in range(2,17):
    url = 'https://ballotpedia.org/Ohio%27s_'+order[i-1]+'_Congressional_District_election,_2020'
    tables = pd.read_html(url)
    this_table = tables[7]
    this_table = this_table[:-1]
    this_table["District"] = i
    data = pd.concat([data,this_table],join='outer',ignore_index=True)
    print("the oder"+ str(i))
    print(this_table)
    time.sleep(6)
    
    
data    
    
# url_1 = 'https://ballotpedia.org/Ohio%27s_11th_Congressional_District_election,_2020'
```
